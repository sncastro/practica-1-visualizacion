## 6. Gráficos a partir de un dataset

- Cargamos el dataset 'titanic' desde seaborn

```python
df = sns.load_dataset('titanic')
# sns.get_dataset_names()
```

```python
df
```

- Indicar a matplotlib que queremos crear una figura con 4 gráficos (2x2)
    1. Gráfico de línea: edad vs. tarifa promedio (age, fare)
    2. Gráfico de barras para la tasa de supervivencia en función de la clase (pclass, survived)
    3. Gráfico de dispersión edad vs. tarifa
    4. Gráfico de pastel Supervivientes vs. No supervivientes (survieved)

```python
age_fare = df.groupby(['age'])['fare'].mean()
survival_class = df.groupby(['pclass'])['survived'].mean()
survived_counts = df['survived'].value_counts()
```

```python
# Definimos los subplots
fig, ax = plt.subplots(2, 2, figsize=(13, 8))

# Primer gráfico
ax[0, 0].plot(age_fare.index, age_fare.values)
ax[0, 0].set_title('Edad vs tarifa promedio')
ax[0, 0].set_xlabel('Edad')
ax[0, 0].set_ylabel('Tarifa promedio')

# 2do
ax[0, 1].bar(survival_class.index,
             survival_class.values,
             tick_label=['1ra clase', '2da clase', '3ra clase'])
ax[0, 1].set_title('Supervivencia por Clase')
ax[0, 1].set_xlabel('Clase')
ax[0, 1].set_ylabel('Proporción de Supervivencia')

# Subplot 3: Gráfico de dispersión edad vs. tarifa
ax[1, 0].scatter(df['age'], df['fare'])
ax[1, 0].set_title('Edad vs. Tarifa')
ax[1, 0].set_xlabel('Edad')
ax[1, 0].set_ylabel('Tarifa')

# Subplot 4: Gráfico pastel Supervivientes vs. No supervivientes
ax[1, 1].pie(survived_counts,
            labels=['No sobrevivientes', 'Sobrevivientes'],
            autopct=lambda x: f'{x:1.2f} %')  # autopct=lambda pct: f'{pct:1.1f} %'  con formato viejo: '%1.1f%%'

plt.tight_layout()
plt.show()
```
